```python
def limpiar_columnas_basura(df):
    cols_to_drop = [c for c in df.columns if 'Unnamed' in c]
    if cols_to_drop:
        print(f"🗑️ Eliminando columnas basura: {cols_to_drop}")
        df.drop(columns=cols_to_drop, inplace=True)
    return df

df_prod = limpiar_columnas_basura(df_prod)
df_socio = limpiar_columnas_basura(df_socio)
df_comm = limpiar_columnas_basura(df_comm)
df_sales = limpiar_columnas_basura(df_sales)
df_prod_desc = limpiar_columnas_basura(df_prod_desc)
```

```python
# SQL Equivalente:
# SELECT s.*, d.product_desc
# FROM sales s
# LEFT JOIN product_description d ON s.product_ID = d.pk_product_ID;

# En Pandas:
df_ventas_enriquecida = pd.merge(
    df_sales,
    df_prod_desc,
    left_on='product_ID',
    right_on='pk_product_ID',
    how='left'
)
df_ventas_enriquecida.head(20)
```

```python
# SQL Equivalente:
# SELECT * FROM sociodemo s
# JOIN commercial c ON s.pk_cid = c.pk_cid AND s.pk_partition = c.pk_partition

# En Pandas:
df_master = pd.merge(
    df_socio,
    df_comm,
    on=['pk_cid', 'pk_partition'],
    how='inner',
    validate='1:1'
)
df_master.head(20)
```

```python
df_master_t = df_master.sort_values(by=['pk_cid', 'pk_partition'])
df_master_t['country_id'] = df_master['country_id'].mask(df_master['country_id'] != 'ES', 'Other')
df_master_t['region_code'] = df_master['region_code'].mask(df_master_t['country_id'] == 'Other', -1.0)
```

```python
# 1. Definimos las columnas
group_cols = ['segment', 'region_code']
target_col = 'salary'

# 2. Calculamos el valor de relleno por grupo (usamos mediana por ser salario)
fill_values = df_master_t.groupby(group_cols)[target_col].median().reset_index()
fill_values.rename(columns={target_col: 'fill_value'}, inplace=True)

# 3. Calculamos el valor global de rescate (por si algún grupo entero es NaN)
global_fallback = df_master_t[target_col].median()

print(f"Mediana global de rescate: {global_fallback}")
print(fill_values.head())

# 4. Hacemos el cruce (merge) del dataframe original con los valores de relleno
df_merged = df_master_t.merge(fill_values, on=group_cols, how='left')

# 5. Rellenamos los nulos directamente en nuestro dataframe original
# Paso A: Rellenamos con el valor de su grupo (segment + region_code)
df_master_t[target_col] = df_master_t[target_col].fillna(df_merged['fill_value'])

# Paso B: Si algún cliente tenía un segmento/región sin ningún salario conocido, usamos el global
df_master_t[target_col] = df_master_t[target_col].fillna(global_fallback)

# Comprobación final
nulos_restantes = df_master_t[target_col].isnull().sum()
print(f"Nulos restantes en {target_col}: {nulos_restantes}")
```

```python
def segment_province(province):
    Provincias_principales = ['Madrid', 'Barcelona', 'Valencia', 'Murcia', 'Sevilla', 'A Coruña']
    Provincias_en_crecimiento = ['Málaga','Pontevedra', 'Alicante', 'Cádiz', 'Asturias', 'Zaragoza', 'Badajoz', 'Las Palmas', 'Valladolid', 'Granada', 'Toledo']

    if province in Provincias_principales:
        return 'Provincias principales'
    elif province in Provincias_en_crecimiento:
        return 'Provincias en crecimiento'
    else:
        return 'Provincias más pequeñas'

df_master_t['province_segment'] = df_master_t['province'].apply(segment_province)
```

```python
# -----------------------------
# Definir variables numéricas
# -----------------------------
numerical_features = [
    'age',
    'salary',
    'months_since_entry',
    'net_margin_sum',
    'num_sales',

    # Variables binarias de productos
    'payroll',
    'pension_plan',
    'payroll_account',
    'emc_account',
    'debit_card',

    # Variables binarias adicionales
    'active_customer'
]


# -----------------------------
# Definir variables categóricas
# -----------------------------
categorical_features = [
    'segment',
    'province_segment'
]


# -----------------------------
# Unir todas las features
# -----------------------------
features_for_clustering = numerical_features + categorical_features

# -----------------------------
# Mostrar información
# -----------------------------
print(f"Cantidad de características numéricas: {len(numerical_features)}")
print(f"Cantidad de características categóricas: {len(categorical_features)}")
print(f"Total de características seleccionadas: {len(features_for_clustering)}")

print("\nPrimeras 5 filas del DataFrame X para clustering:")
display(X.head())
```

```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.impute import KNNImputer

# -----------------------------
# Pipeline de preprocesado
# -----------------------------

numerical_transformer = Pipeline(steps=[
    ('imputer', KNNImputer(n_neighbors=5)),
    ('scaler', StandardScaler())
])

categorical_transformer = Pipeline(steps=[
    ('onehot', OneHotEncoder(handle_unknown='ignore', sparse_output=False))
])

preprocessor = ColumnTransformer(
    transformers=[
        ('num', numerical_transformer, numerical_features),
        ('cat', categorical_transformer, categorical_features)
    ]
)

clustering_pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('kmeans', KMeans(random_state=42, n_init=30))
])

print("Pipeline de clustering definido con éxito:")
print(clustering_pipeline)
```

```python
# -----------------------------
# Entrenamiento final
# -----------------------------
k_optimal = 4

# Actualizar número de clusters de forma segura
clustering_pipeline.set_params(kmeans__n_clusters=k_optimal)

# Entrenar pipeline completo
clustering_pipeline.fit(X)

# Obtener etiquetas
cluster_labels = clustering_pipeline.predict(X)

# Añadir clusters al dataset
X_with_clusters = X.copy()
X_with_clusters['cluster'] = cluster_labels

print(f"Clusters asignados a {X_with_clusters.shape[0]} clientes con k={k_optimal}.")
display(X_with_clusters.head())
```

```python
# -----------------------------
# Perfilado de clusters
# -----------------------------

# Tamaño de cada cluster
cluster_sizes = (
    X_with_clusters['cluster']
    .value_counts()
    .sort_index()
    .to_frame(name='count_customers')
)

# Variables que se agregan por MEDIANA
median_vars = [
    'age',
    'months_since_entry',
    'num_sales'
]

median_df = (
    X_with_clusters
    .groupby('cluster')[median_vars]
    .median()
)

# Variables que se agregan por MODA
mode_vars = [
    'active_customer',
    'segment',
]

mode_df = (
    X_with_clusters
    .groupby('cluster')[mode_vars]
    .agg(lambda x: x.mode().iloc[0] if not x.mode().empty else None)
)


# Variables que se agregan por MEDIA
mean_vars = ['salary', 'net_margin_sum', 'pension_plan', 'payroll_account', 'emc_account', 'debit_card','payroll']

means_df = (
    X_with_clusters
    .groupby('cluster')[mean_vars]
    .mean()
)

means_df.columns = [f"{col}_mean" for col in means_df.columns]

# Unión final
cluster_profile_df = pd.concat(
    [cluster_sizes, median_df, mode_df, means_df],
    axis=1
)

# Ordenar columnas
desired_order = [
    'count_customers',
    'age',
    'months_since_entry',
    'payroll_mean', 'pension_plan_mean', 'payroll_account_mean', 'emc_account_mean', 'debit_card_mean',
    'active_customer',
    'segment',
    'num_sales',
    'salary_mean',
    'net_margin_sum_mean'
]

cluster_profile_df = cluster_profile_df[desired_order]

cluster_profile_df
```

```python
# Mapeo de nombres de clusters
cluster_name_map = {
    0: "Jóvenes universitarios con baja vinculación",
    1: "Clientes premium altamente rentables",
    2: "Adultos con baja vinculación y potencial de crecimiento",
    3: "Clientes vinculados con alta adopción de productos"
}

# Asegurar que el índice sea único y mapear nombres de clusters
cluster_profile_df = cluster_profile_df.reset_index(drop=True)  # reset para evitar duplicados
cluster_profile_df.index = cluster_profile_df.index.map(lambda x: cluster_name_map.get(x, f"Cluster {x}"))
cluster_profile_df.index.name = 'Cluster Name'

# Seleccionar columnas numéricas para el heatmap
numerical_cols_for_heatmap = cluster_profile_df.select_dtypes(include=np.number).columns

# Aplicar estilo con gradiente y formato
styled_cluster_profile = (
    cluster_profile_df
    .style
    .background_gradient(cmap='Blues', subset=numerical_cols_for_heatmap)
    .format(precision=2)
)

# Mostrar el DataFrame estilizado
display(styled_cluster_profile)
```

```python
# A continuación haré un análisis de PCA.

from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.impute import KNNImputer
from sklearn.decomposition import PCA

# -----------------------------
# Transformadores
# -----------------------------
numerical_transformer = Pipeline(steps=[
    ('imputer', KNNImputer(n_neighbors=5)),
    ('scaler', StandardScaler())
])

categorical_transformer = Pipeline(steps=[
    ('onehot', OneHotEncoder(handle_unknown='ignore', sparse_output=False))
])

# -----------------------------
# Preprocessor
# -----------------------------
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numerical_transformer, numerical_features),
        ('cat', categorical_transformer, categorical_features)
    ]
)

# -----------------------------
# Pipeline final con PCA
# -----------------------------
clustering_pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('pca', PCA(n_components=0.95, random_state=42)),
    ('kmeans', KMeans(random_state=42, n_init=30))
])

print("Pipeline de clustering definido con éxito:")
print(clustering_pipeline)
```

```python
# -----------------------------
# Entrenamiento final con PCA
# -----------------------------
k_optimal = 4

# Actualizar número de clusters dentro del pipeline
clustering_pipeline.set_params(kmeans__n_clusters=k_optimal)

# Entrenar pipeline completo (preprocesado → PCA → KMeans)
clustering_pipeline.fit(X)

# Obtener etiquetas de cluster
cluster_labels = clustering_pipeline.predict(X)

# Añadir clusters al dataset original
X_with_clusters = X.copy()
X_with_clusters['cluster'] = cluster_labels

print(f"Clusters asignados a {X_with_clusters.shape[0]} clientes con k={k_optimal}.")
display(X_with_clusters.head())
```

```python
# -----------------------------
# Perfilado de clusters (compatible con PCA)
# -----------------------------


# -----------------------------
# Tamaño de cada cluster
# -----------------------------
cluster_sizes = (
    X_with_clusters['cluster']
    .value_counts()
    .sort_index()
    .to_frame(name='count_customers')
)

# -----------------------------
# Variables agregadas por MEDIANA
# -----------------------------
median_vars = ['age', 'months_since_entry', 'num_sales']

median_df = (
    X_with_clusters
    .groupby('cluster')[median_vars]
    .median()
)

# -----------------------------
# Variables agregadas por MODA
# -----------------------------
mode_vars = ['active_customer', 'segment']

mode_df = (
    X_with_clusters
    .groupby('cluster')[mode_vars]
    .agg(lambda x: x.mode().iloc[0] if not x.mode().empty else None)
)

# -----------------------------
# Variables agregadas por MEDIA
# -----------------------------
mean_vars = ['salary', 'net_margin_sum', 'pension_plan', 'payroll_account', 'emc_account', 'debit_card', 'payroll']

means_df = (
    X_with_clusters
    .groupby('cluster')[mean_vars]
    .mean()
)

# Renombrar columnas de media
means_df.columns = [f"{col}_mean" for col in means_df.columns]

# -----------------------------
# Unión final
# -----------------------------
cluster_profile_df = pd.concat([cluster_sizes, median_df, mode_df, means_df], axis=1)

# -----------------------------
# Ordenar columnas
# -----------------------------
desired_order = [
    'count_customers',
    'age',
    'months_since_entry',
    'payroll_mean', 'pension_plan_mean', 'payroll_account_mean', 'emc_account_mean', 'debit_card_mean',
    'active_customer',
    'segment',
    'num_sales',
    'salary_mean',
    'net_margin_sum_mean'
]

cluster_profile_df = cluster_profile_df[desired_order]

# -----------------------------
# Mostrar el DataFrame final
# -----------------------------
cluster_profile_df
```

## 2. Solución Aplicada: Clustering de Clientes con Preparación y Transformación de Datos

La solución implementada se enfoca en la segmentación de clientes mediante técnicas de clustering, combinando información sociodemográfica, comercial y de productos. El proceso se ha resuelto en varias etapas:

1.  **Integración y Limpieza de Datos**: Se consolidaron múltiples fuentes de datos (sociodemográficos, actividad comercial, productos y ventas) en un único *DataFrame* maestro (`df_master_t`). 3.  **Clustering de Clientes (KMeans)**: Se aplicó el algoritmo de K-Means para agrupar a los clientes en segmentos homogéneos. Antes del clustering, los datos fueron preprocesados utilizando:

Aunque no se logró una reducción significativa en el número de componentes manteniendo este umbral, se incluyó en el pipeline para evaluar su impacto.
5.  **Perfilado de Clusters**: Cada clúster resultante fue perfilado detalladamente, analizando las características medias, medianas y modas de las variables originales. Esto permitió asignar nombres descriptivos a cada segmento (e.g., "Jóvenes universitarios con baja vinculación", "Clientes premium altamente rentables"), facilitando la comprensión y la toma de decisiones estratégicas.
